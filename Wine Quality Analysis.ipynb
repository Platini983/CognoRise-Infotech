```python
## Individual quality plot

fig, axes = plt.subplots(2,3, figsize =(20,15))

axes[0,0].set_title('Quality = 3')
quality_df = wine_df[wine_df.quality==3]
sns.scatterplot(x='free_sulfur_dioxide', y='total_sulfur_dioxide', data = quality_df, s=200, ax=axes[0,0]);

axes[0,1].set_title('Quality = 4')
quality_df = wine_df[wine_df.quality==4]
sns.scatterplot(x='free_sulfur_dioxide', y='total_sulfur_dioxide', data = quality_df, s=200, ax=axes[0,1]);

axes[0,2].set_title('Quality = 5')
quality_df = wine_df[wine_df.quality==5]
sns.scatterplot(x='free_sulfur_dioxide', y='total_sulfur_dioxide', data = quality_df, s=200, ax=axes[0,2]);

axes[1,0].set_title('Quality = 6')
quality_df = wine_df[wine_df.quality==6]
sns.scatterplot(x='free_sulfur_dioxide', y='total_sulfur_dioxide', data = quality_df, s=200, ax=axes[1,0]);

axes[1,1].set_title('Quality = 7')
quality_df = wine_df[wine_df.quality==7]
sns.scatterplot(x='free_sulfur_dioxide', y='total_sulfur_dioxide', data = quality_df, s=200, ax=axes[1,1]);

axes[1,2].set_title('Quality = 8')
quality_df = wine_df[wine_df.quality==8]
sns.scatterplot(x='free_sulfur_dioxide', y='total_sulfur_dioxide', data = quality_df, s=200, ax=axes[1,2]);
```

Except for wine quality 3 and 8, total sulfur dioxide increases as free sulfur dioxide increases, ignoring the outliers.

### QUESTION AND ANSWER

#### Q1. What is the percentage of fixed sulfur dioxide in red wine?

```python
free_sulfur_dioxide_percent = (wine_df.free_sulfur_dioxide)/wine_df.total_sulfur_dioxide
free_sulfur_dioxide_percent.mean()
```

```python
fixed_sulfur_dioxide = (wine_df.total_sulfur_dioxide - wine_df.free_sulfur_dioxide) / wine_df.total_sulfur_dioxide
fixed_sulfur_dioxide.mean()
```

```python
sns.histplot(fixed_sulfur_dioxide)
```
